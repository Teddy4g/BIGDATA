```python
# --- training_script.py ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# --- CONFIGURATION ---
FILE_PATH = 'combined_prices_with_date_cleaned.csv'
STOCKS_TO_PROCESS = ['META_Price', 'TESLA_Price', 'MICROSOFT_Price', 'ORACLE_Price']
LOOK_BACK_DAYS = 60
TRAIN_SIZE_PERCENTAGE = 0.8
EPOCHS = 50
BATCH_SIZE = 32

def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)

def train_and_save_model(df, stock_column_name, look_back, train_size, epochs, batch_size):
    print(f"\n Starting Training for: {stock_column_name}")
    
    # 1. Data Preparation
    price_data = df[[stock_column_name]].copy()
    price_data[stock_column_name] = pd.to_numeric(price_data[stock_column_name], errors='coerce')
    price_data.dropna(inplace=True)
    if price_data.empty:
        print(f" No valid data for {stock_column_name}. Skipping...")
        return

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(price_data.values)
    
    train_data_len = int(len(scaled_data) * train_size)
    train_data = scaled_data[0:train_data_len, :]
    test_data = scaled_data[train_data_len - look_back:, :]
    
    X_train, y_train = create_dataset(train_data, look_back)
    X_test, y_test = create_dataset(test_data, look_back)
    
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # 2. Build LSTM Model
    model = Sequential([
        LSTM(units=50, return_sequences=True, input_shape=(look_back, 1)),
        Dropout(0.2),
        LSTM(units=50, return_sequences=False),
        Dropout(0.2),
        Dense(units=1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # 3. Train Model
    print(f"   Training model for {epochs} epochs...")
    history = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test), verbose=1)
    
    # --- NEW ADDITION: LOSS VISUALIZATION AFTER TRAINING ---
    print(f"   Plotting Training vs Validation Loss...")
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'Loss Curve - {stock_column_name}', fontsize=16) 
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss (Mean Squared Error)', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()
    # ----------------------------------------------------

    # 4. Save Model
    model_path = f'model_{stock_column_name}.h5'
    model.save(model_path)
    print(f" Model for {stock_column_name} has been saved to {model_path}")

# --- MAIN SCRIPT ---
if __name__ == "__main__":
    try:
        main_df = pd.read_csv(FILE_PATH)
        main_df.columns = main_df.columns.str.strip()
        main_df['Date'] = pd.to_datetime(main_df['Date'])
        main_df.set_index('Date', inplace=True)
        print("Main data loaded successfully.")
    except FileNotFoundError:
        print(f" Error: Data file '{FILE_PATH}' not found. Please ensure the file is in the same folder.")
        exit()

    for stock_name in STOCKS_TO_PROCESS:
        train_and_save_model(main_df, stock_name, LOOK_BACK_DAYS, TRAIN_SIZE_PERCENTAGE, EPOCHS, BATCH_SIZE)
    
    print("\n All training processes are complete. ")
```

```python
# --- prediction_and_analysis_script.py ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from tensorflow.keras.models import load_model
import shap
import os

# --- CONFIGURATION  ---
FILE_PATH = 'combined_prices_with_date_cleaned.csv'
STOCKS_TO_PROCESS = ['META_Price', 'TESLA_Price', 'MICROSOFT_Price', 'ORACLE_Price']
LOOK_BACK_DAYS = 60
TRAIN_SIZE_PERCENTAGE = 0.8
# --------------------


def analyze_saved_model(df, stock_column_name, look_back, train_size):
    """
    Function to load a trained model and run it for analysis.
    """
    print("\n" + "="*80)
    print(f" STARTING ANALYSIS FOR STOCK: {stock_column_name}")
    print("="*80)

    # STEP 1: LOAD MODEL
    model_path = f'model_{stock_column_name}.h5'
    if not os.path.exists(model_path):
        print(f" Error: Model file '{model_path}' not found. Skipping this stock.")
        return
    model = load_model(model_path)
    print(f" Model for {stock_column_name} loaded successfully from {model_path}")

    # STEP 2: DATA PREPARATION
    price_data = df[[stock_column_name]].copy()
    price_data[stock_column_name] = pd.to_numeric(price_data[stock_column_name], errors='coerce')
    price_data.dropna(inplace=True)
    if price_data.empty: return

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(price_data.values)
    train_data_len = int(len(scaled_data) * train_size)
    train_data = scaled_data[0:train_data_len, :]
    test_data = scaled_data[train_data_len - look_back:, :]
    
    def create_dataset(dataset, time_step=1):
        dataX, dataY = [], []
        for i in range(len(dataset) - time_step - 1):
            a = dataset[i:(i + time_step), 0]
            dataX.append(a)
            dataY.append(dataset[i + time_step, 0])
        return np.array(dataX), np.array(dataY)

    X_train, y_train = create_dataset(train_data, look_back)
    X_test, y_test = create_dataset(test_data, look_back)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # STEP 3: EVALUATE MODEL
    print(f"\n[INFO - {stock_column_name}] Evaluating model...")
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(predictions)
    y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

    rmse = np.sqrt(mean_squared_error(y_test_actual, predictions))
    mape = mean_absolute_percentage_error(y_test_actual, predictions) * 100
    accuracy_percentage = 100 - mape
    print(f"  - Root Mean Squared Error (RMSE): {rmse:.4f} USD")
    print(f"  - Prediction Accuracy (100% - MAPE): {accuracy_percentage:.2f}%")


    # --- STEP 4: XAI ANALYSIS WITH SHAP ---
    print(f"\n[INFO - {stock_column_name}] Running XAI analysis with SHAP...")

    def predict_wrapper(X):
        X_reshaped = X.reshape(X.shape[0], X.shape[1], 1)
        return model.predict(X_reshaped)

    X_train_2d = X_train.reshape(X_train.shape[0], -1)
    background_summary = shap.kmeans(X_train_2d, 10)
    explainer = shap.KernelExplainer(predict_wrapper, background_summary)

    test_samples_for_shap = 5
    if X_test.shape[0] > test_samples_for_shap:
        X_test_sample = X_test[:test_samples_for_shap]
    else:
        X_test_sample = X_test
    
    X_test_sample_2d = X_test_sample.reshape(X_test_sample.shape[0], -1)

    print(f"[INFO - {stock_column_name}] Calculating SHAP values for {len(X_test_sample_2d)} samples. This may take a while...")
    shap_values = explainer.shap_values(X_test_sample_2d)
    print(f" [INFO - {stock_column_name}] SHAP values calculation complete.")

    feature_names = [f't-{i}' for i in range(look_back-1, -1, -1)]
    
    plt.figure()
    plt.title(f'SHAP Feature Importance - {stock_column_name}', fontsize=16)
    shap.summary_plot(shap_values[0], feature_names=feature_names, plot_type="bar", show=False)
    plt.xlabel("Mean Absolute Impact on Model Output (SHAP Value)", fontsize=12) 
    plt.tight_layout()
    plt.show()
    
    # --- STEP 5: VISUALIZATION & 30-DAY FORECAST ---
    train_plot_data = price_data[:train_data_len]
    valid_plot_data = price_data[train_data_len:].copy()
    predictions_df = pd.DataFrame(predictions, index=valid_plot_data.index[:len(predictions)], columns=['Predictions'])
    valid_plot_data = valid_plot_data.join(predictions_df)
    
    plt.figure(figsize=(16, 8))
    plt.title(f'Prediction vs. Actual - {stock_column_name}', fontsize=18) 
    plt.xlabel('Date', fontsize=14) 
    plt.ylabel('Closing Price (USD)', fontsize=14) 
    plt.plot(train_plot_data[stock_column_name], label='Training Data') 
    plt.plot(valid_plot_data[stock_column_name], color='blue', label='Actual Price (Test)') 
    plt.plot(valid_plot_data['Predictions'], color='red', linestyle='--', label='Predicted Price') 
    plt.legend()
    plt.grid(True)
    plt.show()

    last_days_scaled = scaled_data[-look_back:]
    current_batch = last_days_scaled.reshape(1, look_back, 1)
    future_predictions = []
    for i in range(30):
        current_pred = model.predict(current_batch)[0]
        future_predictions.append(current_pred)
        current_batch = np.append(current_batch[:, 1:, :], [[current_pred]], axis=1)

    future_predictions = scaler.inverse_transform(future_predictions)
    last_date = price_data.index[-1]
    future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=30)
    future_df = pd.DataFrame(future_predictions, index=future_dates, columns=['Predicted_Price'])
    csv_export_path = f'prediksi_30_hari_{stock_column_name}.csv'
    future_df.to_csv(csv_export_path)
    print(f"\n [INFO - {stock_column_name}] 30-day forecast saved to '{csv_export_path}'")

    last_known_price = price_data[stock_column_name].iloc[-1]
    last_known_date = price_data.index[-1]
    last_point_df = pd.DataFrame({'Predicted_Price': [last_known_price]}, index=[last_known_date])
    plot_forecast_df = pd.concat([last_point_df, future_df])
    
    plt.figure(figsize=(16, 8))
    plt.title(f'30-Day Forecast - {stock_column_name}', fontsize=18) 
    plt.xlabel('Date', fontsize=14) 
    plt.ylabel('Closing Price (USD)', fontsize=14) 
    plt.plot(price_data.iloc[-200:], label='Historical Data') 
    plt.plot(plot_forecast_df['Predicted_Price'], color='orange', linestyle='--', label='30-Day Forecast') 
    plt.legend()
    plt.grid(True)
    plt.show()

# --- MAIN SCRIPT ---
if __name__ == "__main__":
    main_df = pd.read_csv(FILE_PATH)
    main_df.columns = main_df.columns.str.strip()
    main_df['Date'] = pd.to_datetime(main_df['Date'])
    main_df.set_index('Date', inplace=True)
    print("Main data loaded successfully.")

    for stock_name in STOCKS_TO_PROCESS:
        analyze_saved_model(main_df, stock_name, LOOK_BACK_DAYS, TRAIN_SIZE_PERCENTAGE)

    print("\n All analysis processes are complete for all stocks. ")
```
